```python
import requests
import time

res = requests.get('http://api.open-notify.org/iss-now.json')

data = []
data_points= 111
for i in range(data_points):
    print(i)
    res = requests.get('http://api.open-notify.org/iss-now.json')
    json_data = res.json()
    data.append(json_data)
    time.sleep(10)
```

```python
data
```

```python
import matplotlib.pyplot as plt

body = []
for details in data:
    row = []
    row.append(details['timestamp'])
    row.append(details['iss_position']['latitude'])
    row.append(details['iss_position']['longitude'])
    body.append(row)

    
latitude = [float(i[1]) for i in body]
longitude = [float(i[2]) for i in body]

plt.figure(figsize=(10, 6))
plt.plot(longitude, latitude, '-o', label='ISS Position', color='pink')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('International Space Station Position')
plt.legend()
plt.grid(True)
plt.show()
```
